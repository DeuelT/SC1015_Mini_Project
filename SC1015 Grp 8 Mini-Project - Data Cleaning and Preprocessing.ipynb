### Exploring '(Kaggle) YouTubeDataset_withChannelElapsed.csv' file 

```python
youtubeData = pd.read_csv('(Kaggle) YouTubeDataset_withChannelElapsed.csv')
pd.set_option('display.max_columns', None)
youtubeData.head()
```

### Exploring '(GitHub) categoryID.csv' file and displaying it

```python
categoryID = pd.read_csv('(GitHub) categoryID.csv')
categoryID = categoryID.drop(columns=['channelID'])
categoryID.head()
```

## 1. Merging two datasets based on category ID to fill in the missing variable of 'Title'

```python
mergedData = pd.merge(youtubeData, categoryID)
pd.set_option('display.max_columns', None)
mergedData.head()
```

```python
mergedData.info()
```

```python
UniqueCount = mergedData["Title"].value_counts().to_dict()
print(UniqueCount)
```

```python
mergedData['GenreCount'] = mergedData["Title"].map(UniqueCount)
```

```python
mergedData[['Title', 'GenreCount']]
```
